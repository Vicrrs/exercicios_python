# Notação Big-O

* Como comparar dois algoritmos?
* Comparação objetiva entre algoritmos
* Considera diferenças entre poder de processamento, sistema operacional, linguaem de programação
* O quanto a "complexidade" do algoritmo aumenta de acordo com as entradas


Usamos a notação big-Θ para limitar de forma assintótica o crescimento do tempo de execução com fatores constantes acima e abaixo. 


A notação Big O é uma notação matemática que descreve o comportamento limitante de uma função quando o argumento tende a um valor ou infinito particular.


Na ciência da computação, a notação Big-O é usada para classificar algoritmos de acordo com a forma como seus requisitos de tempo de execução ou espaço crescem à medida que o tamanho da entrada cresce.

### Função 1 -O(n)

```python
def soma1(n):
    soma = 0
    for i in range(n + 1): # 11 passos
        soma += i
    return soma
```

```python
soma1(10)
```

```python
%timeit soma1(10)
```

### Função 2

```python
def soma2(n):
    return (n * (n+1))/2 # 3 passos O(3)
```

```python
soma2(10)
```

```python
%timeit soma2(10)
```

### Função 3 

```python
def lista1():
    lista = []
    for i in range(1_000):
        lista += [i]
    return lista
```

```python
print(lista1())
```

```python
len(lista1())
```

```python
%timeit lista1()
```

### Funcao 4

```python
def lista2():
    return range(1_000)
```

```python
print(lista2())
```

```python
len(lista2())
```

```python
l = lista2()

for i in l:
    print(i)
```

```python
%timeit lista2()
```

## Funções Big-O
* 1 Constant
* log(n) Logarithmic
* n Linear
* nlog(n) Log Linear
* n^2 Quadratic
* n^3 Cubic
* 2^n Exponential


```python
from math import log
import numpy as np
import matplotlib.pyplot as plt

n = np.linspace(1, 10, 100)
```

```python
len(n)
```

```python
n[0]
```

```python
n[1]
```

```python
n
```

```python
labels = ['Constante', 'Logarithmic', 'Linear', 'Log Linear', 'Quadratic', 'Cubic', 'Exponential']
```

```python
big_o = [np.ones(n.shape), np.log(n), n, n * np.log(n), n**2, n**3, 2**n]
```

```python
plt.figure(figsize=(10, 8))
plt.ylim(0,100)
for i in range(len(big_o)):
    plt.plot(n, big_o[i], label=labels[i])
```

```python
plt.figure(figsize=(10, 8))
plt.ylim(0,100)
for i in range(len(big_o)):
    plt.plot(n, big_o[i], label=labels[i])
plt.legend()
plt.ylabel('Runtime')
plt.xlabel('n')
```

## Exemplos de Big-o

### Constant - O(1)

```python
lista = [1, 2, 3, 4, 5]
```

```python
def constant(n):
    print(n[0])
    print(n[1])
```

```python
constant(lista)
```

### Linear - O(n)

```python
def linear(n):
    for i in n:
        print(i)
```

```python
linear(lista)
```

### Quadratic - O(n^2) - polynomial

```python
def quadratic(n):
    for i in n:
        for j in n:
            print(i, j)
        print('-=-'*5)
```

```python
quadratic(lista)
```

### Combination

```python
# O(1) + O(5) + O(n) + O(n) + O(3)
# O(9) + O(2n) -> O(n)

def combination(n):
    # O(1)
    print(n[0])
    
    # O(5)
    for i in range(5):
        print('test', i)
    
    # O(n)    
    for i in n:
        print(i)
    
    # O(n)    
    for i in n:
        print(i)
        
    # O(3)
    print('Python')
    print('Python')
    print('Python')
```

```python
combination(lista)
```
